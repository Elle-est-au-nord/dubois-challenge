```python
import geopandas
```

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
from matplotlib.colors import ListedColormap
```

```python
data_file = "georgia-1880-county-shapefile/georgia_counties_1870_1880_data.shp"
```

```python
geodata = geopandas.read_file(data_file, driver="shapefile")
```

```python
geodata.head()
```

```python
mpl.rcParams['lines.linewidth'] = 2.0
plt.rcParams['figure.figsize'] = [14, 12] #height, width
# gold, tan, brown, blue, pink, crimson, green
# colorscheme = ["#ffd700", "#d2b48c", "#654321", "#4682b4", "#ffc0cb", "#dc143c", "#00aa00"] #guidelines
colorscheme = ["#fdca41", "#d6c8c3", "#654321", "#042e6a", "#ffa8a8", "#c2255c", "#3d7262"]
dubois_cmap = ListedColormap(colorscheme)
```

```python
geodata.plot(column="data1870 (", cmap=dubois_cmap, edgecolor="k", linewidth=0.2);
geodata.plot(column="data1880_P", cmap=dubois_cmap, edgecolor="k", linewidth=0.2);
```
